# Lets move to CNN Architecture

```python
data = pd.read_csv('./wildlife_drone/mapping_1.csv')
data.head()
```

```python
data.drop(['Unnamed: 0'], axis = 1)
```

```python
data = data.dropna()
```

```python
data = data.drop(['Unnamed: 0'], axis = 1)
```

```python
data.any()
```

```python
X = []
for img_name in data.image_id:
    img = plt.imread('./wildlife_drone/frame1/' + img_name)
    X.append(img)
X = np.array(X)
```

```python
# One Hot Encoder for Categorical Data
y = data.class_id
dummy_y = np_utils.to_categorical(y)
```

```python
image = []
for i in range(0, X.shape[0]):
    a = resize(X[i], preserve_range=True, output_shape=(224, 224)).astype(int)
    image.append(a)
X = np.array(image)
```

```python
from keras.applications.vgg16 import preprocess_input
X = preprocess_input(X, mode='tf')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, dummy_y, test_size=0.3, random_state=0)
```

```python
from keras.models import Sequential
from keras.applications.vgg16 import VGG16
from keras.layers import Dense, InputLayer, Dropout
```

```python
base_model = VGG16(weights = 'imagenet', include_top = False, input_shape = (224, 224, 3))
```

```python
X_train = base_model.predict(X_train)
X_test = base_model.predict(X_test)
X_train.shape, X_test.shape
```

```python
X_train = X_train.reshape(676, 7*7*512)
X_test = X_test.reshape(291, 7*7*512)
```

```python
train = X_train/X_train.max()
X_test = X_test/X_train.max()
```

```python
model = Sequential()
model.add(InputLayer((7*7*512,)))
model.add(Dense(units=1024, activation='sigmoid'))
model.add(Dense(13, activation='softmax'))
```

```python
model.summary()
```

```python
# ii. Compiling the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# iii. Training the model
model.fit(train, y_train, epochs=100, validation_data=(X_test, y_test))
```
